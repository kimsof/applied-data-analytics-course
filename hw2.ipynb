```python
import numpy as np
array = np.round(np.random.rand(100, 100), 3)

print(f'число измерений:      {array.ndim}') 
print('-------------------------------------') 
print(f'размеры массива:      {array.shape}') 
print('-------------------------------------') 
print(f'количество элементов: {array.size}') 
print('-------------------------------------') 
print(f'тип данных элементов: {array.dtype}')
```

```python
import pandas as pd
df = pd.DataFrame(array[0:100:10, 0:100:10])
print('элемент в верхнем левом углу: ', array[-100, -100])
print('элемент в нижнем правом углу: ', array[-10, -10])
df
```

```python
print(f'сумма всех элементов массива: {np.round(np.sum(array))}') 
print('-------------------------------------') 
print(f'среднее арифметическое:       {np.round(np.mean(array), 1)}') 
print('-------------------------------------') 
print(f'стандартное отклонение:       {np.round(np.std(array), 1)}') 
print('-------------------------------------') 
print(f'дисперсия:                    {np.round(np.var(array), 2)}')
print('-------------------------------------') 
print(f'минимум:                      {np.min(array)}')
print('-------------------------------------') 
print(f'максимум:                     {np.max(array)}')
```
