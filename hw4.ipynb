```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
data = np.random.normal(100, 20, 1000)

fig, ax = plt.subplots(3, 1, figsize=(10, 16))
ax[0].plot(x, y, color='green', linestyle='--', linewidth=2)
ax[0].set_title('Линейный график с кастомизацией')
ax[0].set_xlabel('Ось X')
ax[0].set_ylabel('Ось Y')
ax[0].grid()
ax[1].scatter(x, y, c='red', marker='s', alpha=0.5)
ax[1].set_title('Диаграмма рассеяния с кастомизацией')
ax[1].set_xlabel('Ось X')
ax[1].set_ylabel('Ось Y')
ax[1].grid()
ax[2].hist(data, bins=20, color='blue', width=2, edgecolor='blue', alpha=0.5)
ax[2].set_title('Гистограмма с кастомизацией')
ax[2].grid()
```
